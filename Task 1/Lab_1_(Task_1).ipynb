```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
import pandas as pd
```

```python
# random 4x4 array from an exponential distribution
random_array = np.random.exponential(scale=1.0, size=(4, 4))
random_array
```

```python
# random 100000x1 array from the exponential distribution
exp_data = np.random.exponential(scale=1.0, size=100000)
uniform_data = np.random.rand(100000)
normal_data = np.random.normal(size=100000)

plt.hist(exp_data, density=True, bins=50, histtype="step", color="blue", label="Exponential")
plt.hist(uniform_data, density=True, bins=50, histtype="step", color="green", label="Uniform")
plt.hist(normal_data, density=True, bins=50, histtype="step", color="red", label="Normal")
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
# plot Z = X2 + Y2 in a 3D plot with limited x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('Z = X**2 + Y**2')
plt.show()
```

```python
# Loading the CSV File
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
df.head()
```

```python
# visualize the results using heatmaps
corr_pearson = df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].corr(method='pearson')
corr_spearman = df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].corr(method='spearman')

plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
sns.heatmap(corr_pearson, annot=True, cmap='coolwarm', center=0)
plt.title('Pearson Correlation')

plt.subplot(1, 2, 2)
sns.heatmap(corr_spearman, annot=True, cmap='coolwarm', center=0)
plt.title('Spearman Correlation')

plt.tight_layout()
plt.show()
```
